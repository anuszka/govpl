```python
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/new_cases.csv"
# "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/new_deaths.csv"

# "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/new_cases.csv"
# "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/new_cases_per_million.csv"
# image_dir = './images'
# data_dir = './owid_data'
# data_file_cases = 'new_cases_per_million.csv'
```

```python
df=pd.read_csv(url)
```

```python
dfP = df[['date','Poland']].copy(deep=True)
dfP['date'] = pd.to_datetime(dfP['date'], format = "%Y-%m-%d")
# dfP
```

```python
dfP['mean7'] = dfP['Poland'].rolling(7).mean()
# dfP.plot(x='date', y='mean7')
```

```python
dfP2 = dfP[dfP['date'] >= pd.to_datetime(datetime.date(2021,11,1))]
# dfP2.plot(x='date', y='mean7')
```

```python
dfP1 = dfP[(dfP['date'] >= pd.to_datetime(datetime.date(2020,11,1))) & (dfP['date'] <= pd.to_datetime(datetime.date(2021,2,1)))]
# dfP1.plot(x='date', y='mean7')
```

```python
dfP1 = dfP1.reset_index(drop=True)
dfP2 = dfP2.reset_index(drop=True)
dfP1.rename(columns = {'Poland' : 'Poland 2020-2021', 'mean7' : '2020-2021', 'date': 'date 2020-2021'}, inplace=True)
dfP2.rename(columns = {'Poland' : 'Poland 2021-2022', 'mean7' : '2021-2022', 'date': 'date 2021-2022'}, inplace=True)
```

```python
dfp = pd.concat([dfP1, dfP2], axis=1)
dfp['Date'] = dfp['date 2020-2021'].dt.strftime("%d.%m")
# dfp
```

```python
ax = dfp.plot(y=['Poland 2020-2021', 'Poland 2021-2022'])

ticklabels = dfp['Date'].values;

start=0
end=len(ticklabels)
stepsize=3
ax.xaxis.set_ticks(np.arange(start, end, stepsize));

mstepsize=1
ax.xaxis.set_ticks(np.arange(start, end, mstepsize), minor=True);

ax.xaxis.set_ticklabels(ticklabels[::stepsize], rotation=90);
ax.grid(which='major', linestyle='-', linewidth='0.5')
ax.grid(which='minor', linestyle='-', linewidth='0.2')
# ax.set_ylim(0,700);
```
